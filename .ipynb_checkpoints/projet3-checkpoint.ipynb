### data['subject'].unique()

```python
len(data['subject'].unique())
```

```python
data['subject'].value_counts()
```

```python
plt.figure(figsize=(10,8))

ax = sns.countplot(
    x='subject',
    data=data,
    order=data['subject'].value_counts(ascending=False).index,
    palette="Blues_d"   
)

for bars in ax.containers:
    ax.bar_label(bars, fontsize=10, fontweight='bold', color="black")

plt.title("Number of Records per Subject", fontsize=16, fontweight="bold", pad=20)
plt.xlabel("Subject", fontsize=12, fontweight="bold")
plt.ylabel("Count", fontsize=12, fontweight="bold")

plt.xticks(rotation=30, ha="right")

plt.tight_layout()
plt.show()
```

```python
data.head()
```

```python
num_subscribers_per_subject = data.groupby('subject')['num_subscribers'].sum().reset_index().sort_values(by='num_subscribers',ascending=False)
```

```python
num_subscribers_per_subject
```

```python
plt.figure(figsize=(14,8))

ax = sns.barplot(
    data=num_subscribers_per_subject,
    x='subject',
    y='num_subscribers',
    ci=None,
    palette="viridis"   


ax.set_xlabel("Subject", fontsize=14, fontweight="bold")
ax.set_ylabel("Number of Subscribers", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, fontsize=12, ha="right")
plt.yticks(fontsize=12)


plt.title("Subscribers Distribution by Subject", fontsize=18, fontweight="bold", pad=20)


for bars in ax.containers:
    ax.bar_label(bars, fmt="%.0f", fontsize=11, fontweight="bold", color="black")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))

colors = plt.cm.Paired.colors  
plt.pie(
    num_subscribers_per_subject['num_subscribers'],
    labels=num_subscribers_per_subject['subject'],
    autopct='%1.1f%%',
    startangle=140,              
    counterclock=True,
    colors=colors,
    textprops={'fontsize':12, 'fontweight':'bold'}
)

plt.title("Subscribers Percentage by Subject", fontsize=18, fontweight="bold", pad=20)

plt.axis('equal')

plt.show()
```

#  SUBSCRIBERS

```python
data.info()
```

```python
data['published_timestamp'] = data['published_timestamp'].str.split('T').str.get(0)
```

```python
data['published_timestamp'] = data['published_timestamp'].replace("3 hours",'2016-06-21')
```

```python
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'],format='%Y-%m-%d')
```

```python
data['Year'] = data['published_timestamp'].dt.year
data['Month'] = data['published_timestamp'].dt.month
data['Day'] = data['published_timestamp'].dt.day
```

```python
data.head()
```

```python
data.groupby('Year')['num_subscribers'].sum()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("seaborn-v0_8") 

ax = sns.barplot(
    data=data,
    x='Year',
    y='num_subscribers',
    hue='subject',
    ci=None,
    width=0.8,
    palette="tab20"   
)

plt.xlabel("Year", fontsize=14, fontweight="bold")
plt.ylabel("Number of Subscribers", fontsize=14, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Title
plt.title("Number of Subscribers per Year by Subject", fontsize=18, fontweight="bold", pad=20)

for bars in ax.containers:
    ax.bar_label(bars, fmt="%.0f", fontsize=8)

plt.legend(title="Subject", title_fontsize=12, fontsize=11, loc="upper left", bbox_to_anchor=(1,1))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("seaborn-v0_8-whitegrid")  

# Lineplot
sns.lineplot(
    data=data,
    x='Year',
    y='num_subscribers',
    hue='subject',
    ci=None,
    linewidth=2.5,
    marker="o"     
)

# Labels
plt.xlabel("Year", fontsize=14, fontweight="bold")
plt.ylabel("Number of Subscribers", fontsize=14, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Title
plt.title("Trend of Subscribers per Year by Subject", fontsize=18, fontweight="bold", pad=20)

# Legend
plt.legend(title="Subject", title_fontsize=12, fontsize=11, loc="upper left", bbox_to_anchor=(1,1))

plt.tight_layout()
plt.show()
```

# LEVELS

```python
data['level'].value_counts()
```

```python
plt.figure(figsize=(10,8))

ax = sns.countplot(
    x='level',
    data=data,
    palette="Set2"     
)

plt.xlabel("Levels", fontsize=14, fontweight="bold")
plt.ylabel("Count", fontsize=14, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.title("Distribution of Levels", fontsize=18, fontweight="bold", pad=20)

for bars in ax.containers:
    ax.bar_label(bars, color="black", fontsize=11, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))

ax = sns.barplot(data=data,x='level',y='num_subscribers',hue='subject',ci=None)
plt.xlabel("Years",fontsize = 15)
plt.ylabel('Num Of Subscribers',fontsize = 15)
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.style.use("seaborn-v0_8-whitegrid")

#  
num_subjects_level = data.groupby(['subject','level']).size().reset_index(name='count')

#  
ax = sns.barplot(
    data=num_subjects_level,
    x='subject',
    y='count',
    hue='level',
    palette="Set2"
)

#  
plt.xlabel("Subject", fontsize=14, fontweight="bold")
plt.ylabel("Count", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, fontsize=12, ha="right")
plt.yticks(fontsize=12)
plt.title("Distribution of Levels by Subject", fontsize=18, fontweight="bold", pad=20)

# Legend
plt.legend(title="Level", title_fontsize=12, fontsize=11)

plt.tight_layout()
plt.show()
```

```python

def level_charts_per_y(y):
    plt.figure(figsize = (20,10))
    plt.xlabel("Levels",fontsize=15,fontweight='bold',labelpad=20)
    plt.ylabel(f"{y}",fontsize=15,fontweight='bold',labelpad=20)
    plt.title(f'plot of {y} per level')
    ax=sns.barplot(data=data,x='level',y=y,hue='subject',ci=None,width=.7)
    for bars in ax.containers:
        ax.bar_label(bars,fontsize=12,fmt='{:.1f}')
    
    plt.show()
    

    plt.figure(figsize = (10,5))
    data.groupby('level')[y].sum().plot(kind='pie',shadow=True)
    plt.title(f'pie chart of {y} per level')
    plt.show()
```

```python
level_charts_per_y('num_reviews')
```

```python
data['price'].unique()
```

```python
data['price'] = data['price'].replace('Free|TRUE','0',regex=True)
```

```python
data['price'] = data['price'].astype(float)
```

```python
data['Profit'] = data['price'] * data['num_subscribers']
```

```python
plt.figure(figsize=(10,8))
plt.style.use("seaborn-v0_8-whitegrid")

ax = sns.barplot(
    data=data,
    x='Year',
    y='Profit',
    ci=None,
    estimator=sum,
    palette="Blues_r"
)

#  
plt.xlabel("Year", fontsize=14, fontweight="bold")
plt.ylabel("Total Profit", fontsize=14, fontweight="bold")

#  
plt.title("Total Profit by Year", fontsize=18, fontweight="bold", pad=20)

#   
for bars in ax.containers:
    ax.bar_label(bars, fmt="{:.0f}", fontsize=11, fontweight="bold", color="black")

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

# IS_PAID

```python
data["is_paid"].value_counts()
```

```python
paid_dic = {'TRUE':'True','FALSE':'False','https://www.udemy.com/learnguitartoworship/':'True'}
data['is_paid'] = data['is_paid'].map(paid_dic)
```

```python
plt.figure(figsize=(8,6))
plt.style.use("seaborn-v0_8-whitegrid")

#  
ax = data["is_paid"].value_counts().plot(
    kind='bar',
    color=['#2ecc71','#e74c3c'],  #   Paid / Free
    edgecolor="black"
)

# 
plt.xlabel("Course Type", fontsize=14, fontweight="bold")
plt.ylabel("Count", fontsize=14, fontweight="bold")
plt.title("Distribution of Paid vs Free Courses", fontsize=18, fontweight="bold", pad=20)

#    
for bars in ax.containers:
    ax.bar_label(bars, fontsize=12, fontweight="bold", color="black", padding=3)

#    ticks
plt.xticks(ticks=[0,1], labels=["Paid","Free"], fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
